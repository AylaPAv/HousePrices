```python
data_csv = pd.read_csv('data/train.csv')
```

```python
tNullRows = data_csv.isnull().sum().sort_values(ascending = False)
percentNul = (data_csv.isnull().sum()/len(data_csv)*100).sort_values(ascending = False)
missing_data = pd.concat([tNullRows,percentNul],axis = 1, keys = ['Total','Percent'])
missing_data[missing_data['Percent']>6]
```

```python
del_cols = data_csv.columns[data_csv.isnull().sum()/len(data_csv)*100 > 6]
del_cols
```
